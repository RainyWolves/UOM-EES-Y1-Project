# Abstract

Tsunamis have been recorded for many thousands of years and the data is ripe for analysis. This report maps tsunami data to provide a global context for recorded tsunamis, and analyses tsunami intensity from earthquake and volcano derived tsunamis. Tsunamis caused by earthquakes are more prolific, and the data suggests they are somewhat more intense, though further statistical analysis would provide better data. 

## Introduction

Tsunamis are a water-based phenomenon resulting in a wave or series of waves with huge destructive potential. The majority are caused by earthquakes, most of them off-shore, though they can be caused by volcanic activity, landslides, explosions, meteorites, and meteorological activity (Bryant, 2008). The aim of this report is to analyse the locations of recorded tsunamis, taking into account questionable data, and to contrast the intensity of those caused by earthquakes with those caused by volcanoes. One cause of tsunamis registered in the database we will be using for this analysis is deemed as questionable earthquakes. This can be for a number of reasons; before earthquakes were easier identified with seismographs, epicentres were not known, and tsunamis may have been assumed to have been caused by them without adequate enough confirmation. Errors could also have been introduced when data has been taken from varying sources; names of places may have been interpreted incorrectly, and numbers may have been interpreted differently providing different dates or wave heights (World-Wide Tsunamis: 2000 B.C. - to the present, no date). To account for some of these ambiguities, data will be removed from analysis where the cause is determined to have been questionable. The methodology of this report will be via the use of the python programming language, to allow for easy loading, analysing, and visualising of the data.


## Methodology

````python
#### ---------- LOADING THE DATA ---------- ####

# Here we are importing the Pandas package and shortening the term to 'pd' for ease of use as we go along:
import pandas as pd

# The Pandas package will allow us to analyse and structure our data for a much easier experience.
# Next, we are going to use the function 'read_csv' within the Pandas package to import our tsunami data:
data = pd.read_csv("data/tsevent.csv") 

#### ---------- LOADING THE DATA ---------- ####
````

````python
#### ---------- VISUALISING THE DATA ---------- ####

# Here another two pacakages are getting imported; Cartopy and Matplotlib. Both are used to provide visualisation of data.
# Once again they are shortened for ease of use as we continue on.
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# Cartopy and Matplotlib allow us to project the imported data onto a map of Earth, which is what we set up here:
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

# The following code will title the quoted message, providing context to the image:
plt.title("Map 1 - A map of Earth, plotting tsunami locations")

# The following lines extract the latitude and longitude column data values and assign them to variables called 'latitdude'
# and 'longitude', allowing us to plot them onto the map:
Latitude = data['LATITUDE'].values 
Longitude = data['LONGITUDE'].values

# Here we take the coordinate data and plot them to the map we generated earlier, in a format of our choosing.
# The previously defined variables are used, and we choose to plot with a small red dot.
plt.scatter(Longitude, Latitude,
         color='red', linewidth=1, s=9, marker='.',
         transform=ccrs.PlateCarree(),
         )
plt.show()

#### ---------- VISUALISING THE DATA ---------- ####
````

````python
#### ---------- VISUALISING SPECIFIC DATA ---------- ####

# For our second map plot, we only want to display tsunamis that occured after the year 2000 and were not labelled as 
# 'questionable'. Here we create two more plotting variables containing the latitude and longitude data, this time appending
# a '2' to distinguish them from defined in the previous map plot.
Latitude2 = data[(data['YEAR'] > 2000) & (data['CAUSE_CODE'] != 2)]['LATITUDE'].values
Longitude2 = data[(data['YEAR'] > 2000) & (data['CAUSE_CODE'] != 2)]['LONGITUDE'].values

# The nested functions specify that the only latitude and longitude data to be added to the variable should be where the 
# 'YEAR' column has a value over 2000, and the 'CAUSE_CODE' column has any value but '2' which is the questionable earthquake
# code.

# Once again, the following code sets up the map, and we keep the values the same to keep the projections consistent.
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

# The following code will title the quoted message, providing context to the image:
plt.title("Map 2 - A map of Earth, plotting tsunami locations after the year 2000 that were not caused by questionable earthquakes")

# To plot the data onto the map, we use our newly defined variables, and this time we project this data in blue crosses.
plt.scatter(Longitude2, Latitude2,
         color='blue', linewidth=1, s=9, marker='x',
         transform=ccrs.PlateCarree(),
         )
plt.show()

#### ---------- VISUALISING SPECIFIC DATA ---------- ####
````

````python
#### ---------- VISUALISING THE DATA DIFFERENTLY ---------- ####

# The following code will create a histogram of tsunami intensity for those caused only by earthquakes (cause code 1).
# The specified column 'SOLOVIEV' is where intensity is recorded.
data[data['CAUSE_CODE'] == 1].hist(column='SOLOVIEV')

# The x and y axis labels are created with the following commands:
plt.xlabel('Intensity')
plt.ylabel('Number of Tsunamis')

# Finally, a title is created to provide context for the image:
plt.title('Histogram 1 - Displaying the intensity of tsunamis caused by earthquakes')

# The following code will create a histogram of tsunami intensity for those caused only by volcanoes (cause code 6):
data[data['CAUSE_CODE'] == 6].hist(column='SOLOVIEV')

# The x and y axis labels are created with the following commands:
plt.xlabel('Intensity')
plt.ylabel('Number of Tsunamis')

# Finally, a title is created to provide context for the image.
# Matplotlib likes to output the last line as text, to prevent this a semi colon is added to the end:
plt.title('Histogram 2 - Displaying the intensity of tsunamis caused by volcanoes');

#### ---------- VISUALISING THE DATA DIFFERENTLY ---------- ####
````

## Results

Map 1 shows us the locations of all recorded tsunamis. As one would expect, the locations are typically along continental and island coastlines, though a small number of locations are inland. These inland tsunamis are likely to be located in lakes. Map 2 displays the locations of tsunamis recorded after the year 2000 that were not caused by questionable earthquakes. The plotted data is much sparser here as one might expect, though the plotting locations are largely the same. many of the tectonic plate boundaries are represented here as earthquakes and volcanoes are formed, potentially causing tsunamis. Africa displays the starkest difference between Map 1 and Map 2, with only one tsunami registered along the Mediterranean, and one off the coast of South Africa. This could suggest that the data from this region is seen as unreliable, or that there just aren't as many monitoring stations nearby of standard.

Histogram 1 displays the intensity of recorded tsunamis caused by earthquakes and Histogram 2 by volcanoes. The first thing to note is that the number of recorded tsunamis caused by earthquakes is much higher than that of volcanoes, providing more representative data. Both histograms peak at an intensity of around 1, which equates to a maximum runup wave height of around 1.4m. The majority of the results for both causes are over this intensity level, peaking at around 8 for earthquakes, and just under 6 for volcanoes. Two further peaks are noted on both histograms at around an intensity of 3 and 4, which could be due to the nature of the intensity calculation; 5m waves would return an intensity of 2.8, and 15m would return 4.4. The data for wave height is noted to potentially have some discrepancies in recording and converting; if estimated then I believe the peaks could be a result of humans falling back onto numbers that are more familiar to them and easier to work with. It appears that earthquake derived tsunamis have a wider range of intensities but are likely to be more intense than volcanically induced tsunamis.

Tsunami data seems to suffer from having a wide range of inputs and a long history of not-quite-accurate-enough recording. Standardisation of measurements would resolve many discrepancies, though getting different countries and cultures to adopt a unifying system could be difficult. Accurate wave measurements are not easily done quickly enough, and the source of tsunamis is easier now to estimate given seismography, though still not perfect. A future analysis would better be taken with statistical testing in place to determine exact figures around average intensity and to determine which cause would lead to higher intensities.
